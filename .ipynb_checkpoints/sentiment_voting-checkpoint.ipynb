```python
import pandas as pd
import os
```

```python
dirname = os.getcwd()
loc = os.path.join(dirname, "질문답변자 조인")
f1 = pd.read_excel(os.path.join(loc, "학습데이터_추출본_2021_1.xlsx"))
f2 = pd.read_excel(os.path.join(loc, "학습데이터_추출본_2021_2.xlsx"))
f3 = pd.read_excel(os.path.join(loc, "학습데이터_추출본_2021_3.xlsx"))
f4 = pd.read_excel(os.path.join(loc, "학습데이터_추출본_2021_4.xlsx"))
f5 = pd.read_excel(os.path.join(loc, "학습데이터_추출본_2021_5.xlsx"))
li = [f1, f2, f3, f4, f5]
```

```python
df = pd.DataFrame(columns = ['부정', '인사말', '중립', '긍정'])
for i in range(len(f1)):
    tmp = [0, 0, 0, 0]
    for j in range(len(li)):
        tmp[li[j].iloc[i, 9] + 1] += 1
    df = df.append(pd.Series(tmp, index = df.columns), ignore_index = True)
df
```

```python
f = pd.read_excel(os.path.join(loc, "학습데이터_추출본_2021.xlsx"))
result = pd.concat([f, df], axis = 1)
result
```

```python
result.to_excel(os.path.join(loc, "학습데이터_추출본_2021_통합.xlsx"), index = False)
```
