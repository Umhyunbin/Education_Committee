```python
import pandas as pd
import os
import glob
```

```python
def createDirectory(directory): 
    """ 새로운 폴더를 생성하는 함수 """
    try: 
        if not os.path.exists(directory): 
            os.makedirs(directory) 
    except OSError: 
        print("Error: Failed to create the directory.")
```

```python
location = os.getcwd()
dir_loc = os.path.join(location, "회의록_발언_모음")

# 원하는 폴더 이름 설정.
dir_name = "질문답변자 조인"
createDirectory(dir_name)

left = pd.read_csv(os.path.join(dir_loc, "서울특별시의회 회의록 제303회[정례회](2021.11.01 ~ 2021.12.27).csv"), encoding = 'euc-kr')
right = pd.read_csv(os.path.join(dir_loc, "통합본_발언자별 발언수.csv"), encoding = 'euc-kr')

print(left.head())
print(right.head())
```

```python
right = right[right['회수'] == '제303회[정례회](2021.11.01 ~ 2021.12.27)']
right = right[['회수', "발언자", "위원구분"]]
right
```

```python
final = pd.merge(left, right, how = 'outer', on = '발언자')
final
```

```python
final[final['위원구분'].isna()]
```

```python
final['위원구분'] = final['위원구분'].fillna('위원')
```

```python
final.to_csv(os.path.join(dir_name, "303회_조인완료.csv"), encoding = 'euc-kr')
```
